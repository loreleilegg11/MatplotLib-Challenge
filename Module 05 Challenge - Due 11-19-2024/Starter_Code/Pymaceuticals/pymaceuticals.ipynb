```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
study_data_complete = pd.merge(study_results, mouse_metadata, how="left", on="Mouse ID")

# Display the data table for preview
study_data_complete.head(10)
```

```python
# Checking the number of mice.
len(study_data_complete["Mouse ID"].unique())
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mouse = study_data_complete.loc[study_data_complete.duplicated(subset=["Mouse ID","Timepoint"]), "Mouse ID"].unique()
duplicate_mouse[0]
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse_data = study_data_complete.loc[study_data_complete["Mouse ID"] == duplicate_mouse[0]]
duplicate_mouse_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_study_data = study_data_complete[study_data_complete["Mouse ID"].isin(duplicate_mouse)==False]
clean_study_data.head()
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean_study_data["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean of the tumor volume for each regimen
mean_regimen = clean_study_data.groupby("Drug Regimen").mean(numeric_only=True)
mean_regimen["Tumor Volume (mm3)"]
```

```python
# Generate a summary statistics table of medians of the tumor volume for each regimen

median_regimen = clean_study_data.groupby("Drug Regimen").median(numeric_only=True)
median_regimen["Tumor Volume (mm3)"]
```

```python
# Generate a summary statistics table of variance of the tumor volume for each regimen

variance_regimen = clean_study_data.groupby("Drug Regimen").var(numeric_only=True)
variance_regimen["Tumor Volume (mm3)"]
```

```python
# Generate a summary statistics table of standard deviation, of the tumor volume for each regimen

Std_regimen = clean_study_data.groupby("Drug Regimen").std(numeric_only=True)
Std_regimen["Tumor Volume (mm3)"]
```

```python
# Generate a summary statistics table of SEM of the tumor volume for each regimen

Sem_regimen = clean_study_data.groupby("Drug Regimen").sem(numeric_only=True)
Sem_regimen["Tumor Volume (mm3)"]
```

```python
# Assemble the resulting series into a single summary DataFrame.

SummaryDict = {
    "Mean Tumor Volume":mean_regimen["Tumor Volume (mm3)"],
    "Median Tumor Volume": median_regimen["Tumor Volume (mm3)"],
    "Tumor Volume Variance":variance_regimen["Tumor Volume (mm3)"],
    "Tumor Volume Standard Dev.":Std_regimen["Tumor Volume (mm3)"],
    "Tumor Volume Standard Error":Sem_regimen["Tumor Volume (mm3)"]
}
               
df = pd.DataFrame(SummaryDict)
df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
DF2= clean_study_data.groupby("Drug Regimen").agg(
    {"Tumor Volume (mm3)":["mean","median","var", "std", "sem"]}
)
DF2
```

## Quartiles, Outliers and Boxplots

```python

clean_study_data.head()
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# find the greatest time point for the mouse (max), group by mouse id and max vlaue reset index to convert to df
max_time = clean_study_data.groupby("Mouse ID")["Timepoint"].max()
max_time = max_time.reset_index()
max_time.head()
```

```python
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
Data_merge = max_time.merge(clean_study_data, on=["Mouse ID", "Timepoint"], how="left")
Data_merge.head()
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  # Capomulin, Ramicane, Infubinol, and Ceftamin

# Put treatments into a list for for loop (and later for plot labels)
Treatment_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
```

```python
# Create empty list to fill with tumor vol data (for plotting)
tumor_volume_list =[]
```

```python
for treatment in Treatment_list:
    #rows with mice on each drug and tumor volume (.loc())
     information=Data_merge.loc[Data_merge["Drug Regimen"]== treatment]
information
```

```python
# Tumor volumes:
for treatment in Treatment_list:
    #rows with mice on each drug and tumor volume (.loc())
     last_tumor_volume = Data_merge.loc[Data_merge["Drug Regimen"] == treatment, "Tumor Volume (mm3)"]
last_tumor_volume # prints ceftamin
```

```python
# recreate tumor volume
tumor_volume_list =[]
for treatment in Treatment_list:
    #rows with mice on each drug and tumor volume (.loc())
    last_tumor_volume = Data_merge.loc[Data_merge["Drug Regimen"]== treatment, "Tumor Volume (mm3)"]
    
    tumor_volume_list.append(last_tumor_volume)
    
tumor_volume_list
```

```python
 # add subset
tumor_volume_list =[]
for treatment in Treatment_list:
    #rows with mice on each drug and tumor volume (.loc())
    last_tumor_volume = Data_merge.loc[Data_merge["Drug Regimen"] == treatment, "Tumor Volume (mm3)"]
    
    tumor_volume_list.append(last_tumor_volume)
    
# Determine outliers using upper and lower bounds
    quartiles = last_tumor_volume.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upperbound = upperq + (1.5*iqr)
    outliers = last_tumor_volume.loc[(last_tumor_volume < lower_bound) | (last_tumor_volume > upperbound)]
    if len(outliers) == 0:
        print(f"{treatment}'s potential outliers: None")
    else: 
        output = f"{treatment}'s potential outliers:"
        for o in outliers:
            output += f"{o} "
        print(output)
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
# outlier
flierprops = dict(marker="o", markerfacecolor="green", markersize=16, markeredgecolor='black')
#median line
medianprops =dict(linestyle='-',linewidth=2.5, color='green')
#mean customize
meanpointprops = dict(marker="D", markeredgecolor='black', markerfacecolor='green')

plt.boxplot(tumor_volume_list, labels = Treatment_list, flierprops=flierprops,medianprops=medianprops,meanprops=meanpointprops, showmeans=True)
plt.ylabel ('Final Tumor Volume(mm3)')
plt.title("Distribution of the Tumor Volume for Each Treatment Group")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient for mouse weight and average observed tumor volume for the entire Capomulin regimen
correlation = st.pearsonr(capomulinAvgDF["Weight (g)"], capomulinAvgDF["Tumor Volume (mm3)"])[0]
print(f"The correlation between mouse weight and the average tumor volume is {correlation:.2f}")
```

```python
# Calculate a linear regression model for mouse weight and average observed tumor volume for the entire Capomulin regimen
x_values = capomulinAvgDF["Weight (g)"]
y_values = capomulinAvgDF[ "Tumor Volume (mm3)"]

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(capomulinAvgDF["Weight (g)"],
                                                          capomulinAvgDF["Tumor Volume (mm3)"])
regressValues= x_values * slope + intercept
lineEQ = f"y= {slope:.2f}x +{intercept:.2f}"
```

```python
plt.scatter(x_values,y_values, facecolor= "blue", edgecolor ="black", s=40)

plt.plot(x_values, regressValues, "r-")
plt.annotate(lineEQ, (20,70), fontsize=16, color="red")

plt.xlabel('weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title("Average Tumor Volume vs. Mouse Weight for the Capomulin Regimen")
plt.show()
```
